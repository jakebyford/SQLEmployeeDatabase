```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
from config import password

# Create the path to the Departments Database you created in Postgres/SQL
database_path = "../Solved/departments_db.sql"
```

```python
# Create engine as starting point/base layer for Database/Schema. Then create layer 2 "connection"
engine = create_engine(f"postgresql://postgres:{password}@localhost:5432/departments_db")
conn = engine.connect()
```

```python
# Pull the tables and make them readable in Pandas DataFrames
departments = pd.read_sql("Select * from departments", conn)
dept_emp = pd.read_sql("Select * from dept_emp", conn)
dept_manager = pd.read_sql("Select * from dept_manager", conn)
employees = pd.read_sql("Select * from employees", conn)
salaries = pd.read_sql("Select * from salaries", conn)
titles = pd.read_sql("Select * from titles", conn)
```

```python
employees.head()
```

```python
salaries.head()
```

```python
# Merge salaries and emplyees on Employee Number as it is unique to both tables.
employee_salaries = pd.merge(salaries, employees, how = 'left', on = 'emp_no')
# Print new Dataframe
employee_salaries
```

```python
# Create x-values to use as output on the Histogram
x_values = salaries["salary"]
# Create Histogram through Matplotlib
plt.hist(x_values)
plt.title("Salaries")
plt.xlabel("Salary Amount")
plt.ylabel("Number of Common Salaries")
plt.tight_layout()
plt.show()
```

```python
# Find the Average Salaries
average_salary = salaries["salary"].mean()
average_salary
# Rename title_id as emp_title_id making it unique to merge on employee_salaries
titles = titles.rename(columns={"title_id": "emp_title_id"})
titles
```

```python
# Merge Employee Salaries and Titles into Dataframe
avg_salary_by_title_df = pd.merge(employee_salaries, titles, how = 'left', on = 'emp_title_id')
avg_salary_by_title_df

# Group the Avg Salary by Title Dataframe by 'Title'
avg_salary_by_title_group = avg_salary_by_title_df.groupby(['title'])

# Find the mean of the salaries in the new group
avg_salary_by_title_group = avg_salary_by_title_group['salary'].mean()
# Print Group
avg_salary_by_title_group
```

```python
# Create X and Y values - Y are the numbers/X are the Titles
x_values = titles['title'] 
y_values = avg_salary_by_title_group

# Create Bar graph of Average Salaries by Title 
plt.bar(x_values, y_values)
plt.title("Average Salary by Title", fontsize = 20, fontweight = 'bold')
plt.xlabel("Titles")
plt.xticks(rotation = 45)
plt.ylabel("Mean Salary")
plt.tight_layout()
plt.show()
```
